# Step 1: Using Scikit-Learn Module

```python
x = [110,105,115,120,110,130,150,100,105]

import pandas as pd

df = pd.DataFrame(x,columns = ['price'])
df
```

# 1. Normalization / Min Max Scaler using sklearn

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
norm = mms.fit_transform(df[['price']])
norm
```

```python
df['price_norm_sklearn'] = norm
df
```

# 2. Standardization / Standard Scaler using sklearn

```python
from sklearn.preprocessing import StandardScaler 
ss = StandardScaler()
```

```python
df['price_std_sklearn'] = ss.fit_transform(df[['price']])
df
```

# 3. Log Transformation using sklearn

```python
import numpy as np

from sklearn.preprocessing import FunctionTransformer
fc = FunctionTransformer()
```

```python
fc = FunctionTransformer(np.log1p)
df['price_log_sklearn'] = fc.fit_transform(df[['price']])
```

```python
df
```

# 4. Robust Scaler using sklearn

```python
from sklearn.preprocessing import RobustScaler 
RoSc=RobustScaler()
```

```python
df['price_RoSc_sklearn'] = RoSc.fit_transform(df[['price']])
df
```

# 5. Max Absolute Scaler using sklearn

```python
from sklearn.preprocessing import MaxAbsScaler
mx = MaxAbsScaler()
```

```python
df['price_maxAbs_sklearn'] = mx.fit_transform(df[['price']])

df
```

# Step 2: Using Python Manually

```python
import pandas as pd
import numpy as np

df2 = pd.read_excel('E:\\Machine Learning Course\\Assignments\\Assignment 5 (data).xlsx')
df2.head(3)
```

# 1. Normalization / Min Max Scaler using Python Manually

```python
x = df2.price
```

```python
df2['price_norm_manual'] = (x-min(x))/(max(x)-min(x))
df2
```

# 2. Standardization / Standard Scaler using Python Manually

```python
df2['cost_std_manual'] = (x-x.mean())/x.std(ddof=0)
```

```python
df2
```

# 3. Log Transformation using Python Manually

```python
df2['price_log_manual'] = np.log1p(x)
df2
```

# 4. Robust Scaler using Python manually

```python
df2['price_RoSc_manual'] = (x-np.median(x))/(np.percentile(x,75) - np.percentile(x,25))
```

```python
df2
```

# 5. Max Absolute Scaler using Python Manually

```python
df2['price_maxAbs_manual'] = x/max(x)
```
